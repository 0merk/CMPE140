# Section 1 Question

First, import the file [USA_data.csv](./USA_data.csv) into R. 

Then plot **UNEMPLOYMENT** and **INFLATION** on the same graph against TIME.

You will use **plot()** and **lines()** functions.

1. Label the x-axis as *"Years"* and limit x axis by 1960 and 2018.
2. Label the x-axis as *"Unemp & Inf"* and limit y axis by 0 and 15.
3. Set the color for Unemployment to <font color='blue'>blue</font>.
4. Set the color for Inflation to <font color='red'>red</font>.
5. Set the title of the figure as *"Philips Curve"*.

```r
usa_data <- read.csv("USA_data.csv")

plot(usa_data$TIME, usa_data$UNEMPLOYMENT, col="blue", type="l", lwd=3, 
     xlab="Years", ylab="Unemp & Inf", xlim=c(1960,2018), ylim=c(0,15))

lines(usa_data$TIME, usa_data$INFLATION, col="red", type="l", lwd=3)

title("Philips Curve")
```

# Section 2 Question

First, import the file [Sales.csv](./Sales.csv) into R. 

Make a probability density plot of **Quantity** using **hist()** function and add a density curve on this plot by using the **lines()** function
    
1. Set the title of the figure as *"Histogram for Sales"*.
2. Label the x-axis as *"Sales"*.
3. Set the color to <font color='green'>green</font>.

Hint: Use probabilities in your histogram.

Hint: **lines()** function will include **density()** of the relevant data.

```r
sales <- read.csv("Sales.csv")

hist(sales$Quantity, breaks = 15, col = "green", freq = F,
     main = "Histogram for Sales", xlab = "Sales")

lines(density(sales$Quantity))
```
