# Section 1 Question

First, import the file [CGDPo.tsv](./CGDPo.tsv) into R.

Your plot should include three subplots. They should be placed vertically in given order.

1. **Line plot:** showing the changes in merchandise exports **csh_x** with a blue line and residual trade **csh_r** with a <font color='red'>red</font> line over time **year**.
   - Title should be *"csh_x & csh_r"*
   - x-axes should be labeled as *"Years"*
   - y-axes should be labeled as *"Shares"*
2. **Histogram:** Histogram should represent government consumption **csh_g**.
   - Title should be *"Histogram"*
   - x-axes should be labeled as *"Government Consumption"*
   - It should have **7 breaks** and the bars should be colored <font color='green'>green</font>
   - x-axes should be limited by 0.05 and 0.25
3. **Pie chart:** This plot should be the absolute of averages of shares labeled with the name of the header. For example, “csh_m” labeled slice of the pie chart has the value of 0.08966396.
   - Title should be *"Pie Chart"*

```r
CGDPo <- read.table("CGDPo.tsv", dec = ",", header = T)

par(mfrow=c(3,1))

plot(CGDPo$year, CGDPo$csh_x, 
     type="l", 
     col="blue", 
     main="csh_x & csh_r",
     xlab="Years", 
     ylab="Shares")
lines(CGDPo$year, CGDPo$csh_r, col="red")


hist(CGDPo$csh_g, 
     breaks =7,
     col="green", 
     xlab="Government Consumption",
     main="Histogram", 
     xlim=c(0.05,0.25))


total <- abs(apply(CGDPo, 2, mean)[-1])
pie(total, main="Pie Chart")
```

# Section 2 Question

First, import the file [CGDPo.tsv](./CGDPo.tsv) into R.

Your plot should include four subplots in 2×2 shape which are described below in top-to-bottom and left-to-right.

1. **Box Plot:** Plot the government consumption which has the title **csh_g** in the data. This plot should have title *"Box Plot"* above.

2. **Scatter Plot:** This plot should represent household consumption which has the title **csh_c** in the data. Points should be <font color='green'>green</font> circles. It should have title *"Index"* on below, *"Consumption"* on left, and *"Plot"* above.

3. **Histogram:** This plot should also represent the government consumption data. Bars should be colored <font color='red'>red</font>. Bar count should be 4. The plot should be titled *"Government Consumption"* below and *"Histogram"* above.

4. **Strip Plot:** This plot should represent government consumption again. Points should be <font color='blue'>blue</font> circles. The plot should be titled *"Strip Chart"* above.

```r
CGDPo <- read.table("CGDPo.tsv", dec = ",", header = T)

par(mfrow=c(2,2))

boxplot(CGDPo$csh_g, main = "Box Plot")

plot(CGDPo$csh_c, col = "green", xlab = "Index", ylab = "Consumption", main = "Plot")

hist(CGDPo$csh_g, col = "red", breaks = 3,
     xlab = "Government Consumption", main = "Histogram")

stripchart(CGDPo$csh_g, pch = 1, col = "blue", main = "Strip Chart")
```
