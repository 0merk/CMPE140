## 4.1 QUESTION

Sometimes understanding the how functions are related is crucial for data analysis, and the best way to do that is visualising those functions. To achieve this, we want to create a function that shows the relationship between those functions in a _matrix like_ format, where every row and column represent a function.

1. Create a function named **plot_matrix** with 4 arguments:
    + **first_f, second_f, third_f**: functions that we want to visualize
    + **range**: range of values that the parameter value will take, with a default value: `seq(from = 0, to = 2*pi, length.out = 200)`
    
    
2. Inside **plot_matrix**, set the `mfcol` argument of the `par()` function to an appropriate value so that 9 plots can be displayed at the same time (with 3 rows and 3 columns)

3. Create 9 parametric plots over **range** such that:
    + y axis of every plot in the first row and x axis of every plot in the first column is **first_f**
    + y axis of every plot in the second row and x axis of every plot in the second column is **second_f**
    + y axis of every plot in the third row and x axis of every plot in the third column is **third_f**
    
**Example:**
```r
> plot_matrix(sin, sqrt, cos)
```


## 4.2 SOLUTION

````python
#1
plot_matrix <- function(first_f, second_f, third_f, range = seq(from = 0, to = 2*pi, length.out = 200))
{
    #2
    par(mfcol = c(3,3))
    
    #3
    plot(first_f(range), first_f(range), type = "l", ylab = "First Function", xlab = " ", main = "First Function")
    plot(first_f(range), second_f(range), type = "l", ylab = "Second Function", xlab = " ")
    plot(first_f(range), third_f(range), type = "l", ylab = "Third Function", xlab = " ")
    
    plot(second_f(range), first_f(range), type = "l", ylab = "", xlab = " ", main = "Second Function")
    plot(second_f(range), second_f(range), type = "l", ylab = " ", xlab = " ")
    plot(second_f(range), third_f(range), type = "l", ylab = " ", xlab = " ")
    
    plot(third_f(range), first_f(range), type = "l", ylab = "", xlab = " ", main = "Third Function")
    plot(third_f(range), second_f(range), type = "l", ylab = " ", xlab = " ")
    plot(third_f(range), third_f(range), type = "l", ylab = " ", xlab = " ")
    
}

plot_matrix(sin, sqrt, cos)
````

# 7 ASTROID AND CIRCRLE

## 7.1 QUESTION

An astroid is a particular mathematical curve whose cartesian equation is:

$$ x^{2/3} + y^{2/3} = a^{2/3} $$  

And its parametric equation is given as follows:

$$ x(t) = a \times cos^3(t) $$
$$ y(t) = a \times sin^3(t) $$
$$ t \in [0 , 2\pi] $$

where a denotes the radius of a fixed circle. 

1. Create a function called **astroid_and_circle** whose arguments are **x** and **y**, which creates 2 graphs:


- a circle whose radius is **x** and an astroid (the parameter is **a** is **x**) inside that circle


- a circle whose radius is **y** and an astroid (the parameter is **a** is **y**) inside that circle


2. For example:

```r
> astroid_and_circle(2, 3)
```


**Hint**: Set the parameter t using `seq()` function and choose the argument **length.out** as 200.

## 7.2 SOLUTION

````python
#1 and #2

t <- seq(0,2*pi,length.out = 200)

astroid_and_circle <- function(x, y)
{
  par(mfrow = c(1,2))
  plot(x * (cos(t)) ^ 3, x * (sin(t)) ^ 3, type = "l", col = "red")
  points(x * cos(t), x * sin(t), type = "l", col = "blue")
  plot(y * (cos(t)) ^ 3, y * (sin(t)) ^ 3, type = "l", col = "red")
  points(y * cos(t), y * sin(t), type = "l", col = "blue")
}

astroid_and_circle(2, 3)
````

# 15 DOT, BAR, PIE CHARTS

## 15.1 QUESTION

There is a built-in data set in R Studio called *islands*. 

1. Create a function called **pie_chart_maker** that will take a given *x* variable. 


2. This function should first return a rainbow-colored pie chart of the largest 5 variables of the *x* vector called **Largest**.


3. And then return a rainbow-colored pie chart of the smallest 5 variables of the *x* vector called **Smallest**.


4. Use your **pie_chart_maker** on the built-n data set **islands**.


Your final pie charts should look like this:


## 15.2 SOLUTION

````python
#1
pie_chart_maker <- function(x)
{
  
  #2
  largest <- sort(x, decreasing = TRUE)[1:5]
  my_pie_largest <- pie(largest, col = rainbow(length(largest)), main = "Largest")
 
  #3
  smallest <- sort(x, decreasing = FALSE)[1:5]
  my_pie_smallest <- pie(smallest, col = rainbow(length(smallest)), main = "Smallest")
  return(c(my_pie_smallest, my_pie_largest))
}

#4
pie_chart_maker(islands)
````
