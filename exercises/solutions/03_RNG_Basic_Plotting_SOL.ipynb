# THROWING THE DICE, PICKING THE PAPER

## QUESTION

Genar and Ada are playing a game. Each of them throws a dice three times and writes the outcomes seperately onto a piece of paper.

1. Run the following code in RStudio:

```r
RNGversion("3.3.1")

set.seed(2)
```

2. Write down the experiment and put the results into the vectors **genar** and **ada** respectively (Use the `sample()` function and set the argument **replace** as TRUE).


3. Name all elements in these vectors according to each player's name's initial letter ("g" for Genar, "a" for Ada). Example for different seed:

```r
> genar
g g g 
6 5 2 

> ada
a a a 
4 6 6 
```

4. In this step, each player updates his/her vector. Each of them holds the numbers which are greater than the average of the vector (the three number's average) and gets rid of the rest. 


5. After the update in respective vectors, a third person gathers these two vectors into one vector called **all_x**. This vector represents a box containing papers which includes numbers from the vectors **genar** and **ada**.


6. Then a third person selects two papers from this box (selects two numbers from the **all_x** vector without replacement). Then she puts these two numbers into another vector called **choose**.


7. The third person updates the **genar** and **ada** vectors once again. From the **choose** vector, she puts the number coming from Genar's vector into Ada's vector and vice versa.


8. Finally, the numbers in the **genar** and **ada** vectors are summed seperately. The one having the bigger number wins. Create a vector called **winner** and write the winner's name in it. 

## SOLUTION

````r
#1
RNGversion("3.3.1")
set.seed(20)

#2
genar <- sample(1:6, 3, replace = T)
ada <- sample(1:6, 3, replace = T)

#3
names(genar) <- rep("g", 3)
names(ada) <- rep("a", 3)

#4
genar <- genar[genar > mean(genar)]
ada <- ada[ada > mean(ada)]

#5
all_x <- c(genar, ada)

#6
choose <- sample(all_x, 2, replace = F)

#7
genar <- c(genar, choose["a"])
ada <- c(ada, choose["g"])

#8
sum(genar)
sum(ada)
winner <- "Ada"
````

# GRIMLEY'S FAITH 

## QUESTION

Copy this code below in order to get the same result with the solution.
```r
RNGversion("3.3.1")
set.seed(104)
```

Grimley is taking a difficult mathematics course in his College. All of the registered students in this course participated to 3 exams: 2 midterms and a final. In the course syllabus the professor mentioned that their final grade will be evaluated as follows: 

 > "30% of midterm 1, 30% of midterm2 and 40% of final notes. Participation is not mandatory but suggested."

There are 100 students in the class and the requirement to pass this course is to be on or above the curve of final adjusted notes. Grimley is the 45th student in the list. All the notes are according to this list. It is time for him to check whether he passed or not.

1. Create a vector of *100 randomly dstributed numbers, with a mean of 47 and a standard deviation of 19* and name it as: **midterm1**


2. Create another vector of *100 randomly dstributed numbers, with a mean of 54 and a standard deviation of 12* and name it as: **midterm2**


3. Create another vector of *100 randomly dstributed numbers, with a mean of 35 and a standard deviation of 13* and name it as: **final**


4. Adjust the grades according to the percentages mentioned on the syllabus(30% of midterm 1, 30% of midterm2 and 40% of final notes) and name the adjusted vectors as: **midterm1_adj, midterm2_adj, final_adj**


5. Sum these adjusted notes and assign the sum vector named as: **adjustment**


6. Calculate the mean of the adjusted grades and name it: **curve**


7. Subset **adjustment** to find note of adjusted note of the grimley and assign this value to variable named **grimley**. "*Hint: Remember Grimley was the 45th student in the list.*


8. Finally create a variable called **passed_or_not** as follows: result of the check that if Grimley's adjusted grade is greater or equal than the curve.

```r
[1] TRUE
```

## SOLUTION 

````r
#1
RNGversion("3.3.1")
set.seed(104)

#2
midterm1 <- rnorm(n = 100, mean = 47, sd = 19)
midterm2 <- rnorm(n = 100, mean = 54, sd = 12)
final <- rnorm(n = 100, mean = 35, sd = 13)

#3
midterm1_adj <- midterm1 * 0.3
midterm2_adj <- midterm2 * 0.3
final_adj <- final * 0.4

#4
adjustment <- midterm1_adj + midterm2_adj + final_adj

#5
curve <- mean(adjustment)

#6
grimley <- adjustment[45]

#7
passed_or_not <- grimley >= curve
passed_or_not
````
