## Q4

1) Write a funcion that calculates the letter grades of a lecture. 
The input should be a vector with elements midterm 1, midterm 2, and final grades.
The weights are 30%, 30%, and 40% respectively. The intervals for the letter grades are like the following:

100-90 AA / 90-85 BA / 85-80 BB / 80-75 CB / 75-70 CC / 70-65 DC / 65-60 DD / 60-0 F
The function should return the letter grades.

2) The notes for the courses are as follows:

cmpe140 82 - 71 - 90

ec203 98 - 75 - 84

ad231 45- 63 - 72

Create vectors for each lecture and store these vectors into a list called grades_list . 

3) Find the letter grades for each course.

````r
#1
grade_func <- function(grade_vec) {
  mean_grade <- 0.3*grade_vec[1] + 0.3*grade_vec[2] + 0.4*grade_vec[3]
  if(mean_grade >= 90) {
    return("AA")
  } else if (mean_grade >= 85 ){
    return("BA")
  } else if ( mean_grade >= 80 ){
    return("BB")
  } else if (mean_grade >= 75 ){
    return("CB")
  } else if (mean_grade >= 70 ){
    return("CC")
  } else if (mean_grade >= 65 ){
    return("DC")
  } else if (mean_grade >= 60 ){
    return("DD")
  } else {
    return("F")
  }
}

#2
cmpe140 <- c(82,71,90)
ec203 <- c(98,75,84)
ad231 <- c(45,63,72)
grades_list <-list("cmpe"=cmpe140, "micro"=ec203, "law"=ad231)

#3
lapply(grades_list, grade_func)
````

````r
#1
long_trees <- trees[trees$Height > 80, ]

#2
biggest_tree <- function(x){
  a <- c()
  for(i in 1:nrow(x)){
    a <- c(a,sum(x[i,]))
  }
  return(x[which.max(a),])
}

#3
biggest_tree(long_trees)
````
