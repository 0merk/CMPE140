Fixed cost refers to costs that do not depend on the number of units produced such as rent, utilities etc. whereas variable costs are the costs proportional to the amount of production such as raw material costs.

Your function should take three inputs:

- fixed cost
- variable cost
- price per unit

Calculate the break-even point of a company whose fixed cost is 5000 dollars and variable cost is 1.8 dollars. The price of their product is 2.5. 

Don't forget to round the result to the lowest integer which is higher than or equal to it so you don't report a decimal number. For this, you can use the **ceiling()** function.

```r
> BEP(2.5, 5000, 1.8)
[1] 7143
```
