# KNOW YOUR PROFIT

Suppose you are a finance manager and you want to calculate the profit before tax (pbt) of your company using only three values:

    sales * gross margin - operating costs

Create a function named **profit** that takes three arguments:

- **sales**, a numeric value for net sales, with no default value
- **margin**, a numeric value between 0 and 1 that shows gross profit / sales, with no default value
- **costs**, a numeric value for fixed operating costs, **with a default value of 100**

and returns the profit before tax as such:

```R
profit(sales = 1000, margin = 0.3)

[1] 200

profit(sales = 500, margin = 0.2, costs = 40)

[1] 60
```

````r
profit <- function(sales, margin, costs = 100)
{
    return(sales * margin - costs)    
}

profit(sales = 1000, margin = 0.3)

profit(sales = 500, margin = 0.2, costs = 40)
````

# GRADUATE OR NOT, VERSION 2

Suppose in a university there are two conditions to be met in order to graduate:

- **At least** 10 courses must be taken
- The GPA (average grade) must be **at least** 2

Write a function named **grad2** that takes a single argument **grades**  - a vector of grades - and:
- First checks whether the **"at least 10 courses"** condition is **NOT** met, and **RETURNS** **"courses"** if the condition is **NOT** met
- If the first condition is met, then checks whether **"at least 2 gpa"** condition is **NOT** met and **RETURNS** **"gpa"** if the condition is **NOT** met
- **RETURNS** **"yes"** if both conditions are met

```R
grds <- seq(0.5, 4, 0.5)

RNGversion("3.3.1")
set.seed(10)
prbs <- runif(length(grds))
grades1 <- sample(grds, sample(20, 1), replace = T, prob = prbs/sum(prbs))

RNGversion("3.3.1")

grades2 <- sample(grds, sample(20, 1), replace = T, prob = prbs/sum(prbs))
```

**Hint:**
- Make use of a nested **if()... else ...** construct
- You may use **mean** and **length** functions
- You can invert the conditions in order to check whether the condition is **NOT** met (**"less than"** complements **"greater than or equal to"**)

````r
grds <- seq(0.5, 4, 0.5)

RNGversion("3.3.1")
set.seed(10)
prbs <- runif(length(grds))
grades1 <- sample(grds, sample(20, 1), replace = T, prob = prbs/sum(prbs))

RNGversion("3.3.1")
set.seed(500)
prbs <- runif(length(grds))
grades2 <- sample(grds, sample(20, 1), replace = T, prob = prbs/sum(prbs))

RNGversion("3.3.1")
set.seed(150)
prbs <- runif(length(grds))
grades3 <- sample(grds, sample(20, 1), replace = T, prob = prbs/sum(prbs))

grades1
grades2
grades3

grad2 <- function(grades)
{
    gmean <- mean(grades)
    glen <- length(grades)
    
    if (glen < 10)
    {
        return("courses")
    }
    else if (gmean < 2)
    {
        return("gpa")
    }
    else
    {
        return("yes")
    }

}

grad2(grades = grades1)
grad2(grades = grades2)
grad2(grades = grades3)
````
