# KNOW YOUR BREAKEVEN

Suppose you are a finance manager and you want to calculate the breakeven point of your company using only three values:

    operating costs / (price * gross margin)
    
This value shows the number of units to sell in order to cover all (fixed) operating costs 

Create a function named **breakeven** that takes three arguments:

- **price**, a numeric value for the price, with no default value
- **margin**, a numeric value between 0 and 1 that shows gross profit / sales, with no default value
- **costs**, a numeric value for fixed operating costs, **with a default value of 200**

and returns the breakeven point as such:

```R
breakeven(price = 10, margin = 0.2)

[1] 100

breakeven(price = 50, margin = 0.1, costs = 400)

[1] 80
```

````r
breakeven <- function(price, margin, costs = 200)
{
    return(costs / (price * margin))    
}

breakeven(price = 10, margin = 0.2)

breakeven(price = 50, margin = 0.1, costs = 400)
````

# ATHLETE, VERSION 1

An athlete keeps his 100 meters sprint performances in a competition in a vector.  The competition committee grants him a badge of **"good"** if:

- Either any of his performances is less than 9.58 (the world record set by Usain Bolt in 2009)
- **OR** all his performances are less than 10.

He recieves **bad** otherwise

Write a function named **perform1** that takes a single argument **times** and **RETURNS** **"good"** if **EITHER** condition is met and **"bad"** otherwise

**Hint:**
- Make use of **if()... else ...** construct and **|** operator
- The most concise way to check the required conditions is to use **any** and **all** functions

````r
RNGversion("3.3.1")
set.seed(10)
seconds1 <- round(rnorm(10, 9.9, 0.4), 2)

RNGversion("3.3.1")
set.seed(500)
seconds2 <- round(rnorm(10, 9.8, 0.3), 2)

RNGversion("3.3.1")
set.seed(150)
seconds3 <- round(rnorm(10, 9.9, 0.1), 2)

seconds1
seconds2
seconds3

perform1 <- function(times)
{
    if (any(times < 9.58) | all(times < 10))
    {
        return("good")
    }
    else
    {
        return("bad")
    }
}

perform1(times = seconds1)
perform1(times = seconds2)
perform1(times = seconds3)
````

# ATHLETE, VERSION 2

An athlete keeps his 100 meters sprint performances in a competition in a vector.  The competition committee grants him a badge of:

- **"record"** if any of his performances is less than 9.58 (the world record set by Usain Bolt in 2009)
- **"stable"** if all of his performances are less than 10 seconds
- **"bad"** if neither of the above conditions is met

Write a function named **perform2** that takes a single argument **times** and:
- First checks whether any performance is less than 9.58 seconds and RETURNS **"record"** if the condition is satisfied
- Else checks whether all performances are less than 10 seconds and RETURNS **"stable"** if the condition is satisfied
- Else RETURNS **"bad"** if none of the conditions is met

```R
RNGversion("3.3.1")
set.seed(10)
seconds1 <- round(rnorm(10, 9.9, 0.4), 2)

RNGversion("3.3.1")
set.seed(500)
```

**Hint:**
- Make use of a nested **if()... else ...** construct
- The most concise (but not the only) way to check the required conditions is to use **any** and **all** functions

````r
RNGversion("3.3.1")
set.seed(10)
seconds1 <- round(rnorm(10, 9.9, 0.4), 2)

RNGversion("3.3.1")
set.seed(500)
seconds2 <- round(rnorm(10, 9.8, 0.3), 2)

RNGversion("3.3.1")
set.seed(150)
seconds3 <- round(rnorm(10, 9.9, 0.1), 2)

seconds1
seconds2
seconds3

perform2 <- function(times)
{
    if (any(times < 9.58))
    {
        return("record")
    }
    else if (all(times < 10))
    {
        return("stable")
    }
    else
    {
        return("bad")
    }
}

perform2(times = seconds1)
perform2(times = seconds2)
perform2(times = seconds3)
````
